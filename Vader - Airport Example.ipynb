```python
import pandas as pd
data = pd.read_csv("D:/4th year/Grad Project/Datasets/Airline-Sentiment-2-w-AA.csv" , encoding = "ISO-8859-1")
```

```python
#Isolate Tweet Texts
tweets = data["text"]
#Remove @'s to Make Sentiment Analysis more Effective
import re
def remove_ats(tweet):
 """Removes the @ symbol and the word following it from a 
tweet"""
 return re.sub(r"@\w+", "", tweet)
tweets_without_at = [remove_ats(text) for text in tweets]
```

```python
import nltk
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Create sentiment scores for each tweet
nltk.download("vader_lexicon")  # Download the sentiment lexicon
analyzer = SentimentIntensityAnalyzer()

sentiments = []
for tweet in tweets_without_at:
    sentiments.append(analyzer.polarity_scores(tweet).get("compound"))
data["sentiment"] = sentiments

```

```python
#Visualize sentiment over time by airline
def visualize_sentiment_over_time(airline):
    """Visualizes the compound sentiment scores of tweets about different airlines over time by hour"""
    airline_data = data[data['airline'] == airline]
    airline_data['tweet_created'] = pd.to_datetime(airline_data['tweet_created'])
    airline_data.set_index('tweet_created', inplace=True)
    airline_data_hourly = airline_data.resample('H').mean(numeric_only=True)
    
    # Set the figure size and title
    plt.figure(figsize=(10, 6))
    plt.title("Compound Sentiment Scores Over Time For " + airline)
    plt.ylim([-1, 1])
    
    # Plot the compound sentiment scores as a line chart
    plt.plot(airline_data_hourly.index.values, airline_data_hourly["sentiment"])
    
    # Add a legend, axis labels, and show the plot
    plt.legend()
    plt.xlabel("Time Tweeted")
    plt.ylabel("Compound Sentiment Score")
    plt.show()
    
    return(airline_data_hourly)
    
va_hourly = visualize_sentiment_over_time("Virgin America")
united_hourly = visualize_sentiment_over_time("United")
sw_hourly = visualize_sentiment_over_time("Southwest")
delta_hourly = visualize_sentiment_over_time("Delta")
usa_hourly = visualize_sentiment_over_time("US Airways")
american_hourly = visualize_sentiment_over_time("American")
```

```python
#What are the main indicators of very negative or damaging tweets?
#Negative tweets categorized as tweets with a sentiment score under -.5
negative_tweets = data[data['sentiment'] < -0.5]['text']
#What are the most common issues between the 4 categories mostly seen in the negative tweet data?
#1. Flight Cancellations and Delays
flight_indicator_words = ['delay', 'delayed', 'cancel', 'cancelled','changes','change','stuck', 'reschedule', 'rescheduled']
#2. Price and Reimbursement Issues
money_indicator_words = ['price','$','$$','$$$','money','reimbursement','reimburse','pay','paid']
#3. Customer Service Issues
customer_service_indicator_words = ['customer service','attitude','staff','rude','employees','employee','unwilling','inexperienced','complaints','complaint', 'agent','agents']
#4. Lost Baggage or Poor Handling
baggage_indicator_words = ['bag','bags','luggage','baggage','handling','lost','cargo','handlers','broke',]
#Set counters for each issue
num_flight = 0
num_money = 0
num_cs = 0
num_bag = 0
num_total = len(negative_tweets)
#Categorize each negative tweet
for tweet in negative_tweets:
    if any(word in tweet for word in flight_indicator_words):
        num_flight += 1
    if any(word in tweet for word in money_indicator_words):
        num_money += 1
    if any(word in tweet for word in customer_service_indicator_words):
        num_cs += 1
    if any(word in tweet for word in baggage_indicator_words):
        num_bag += 1
#Plot % of Total Negative Tweets by Issue
percent_flight = (num_flight/float(num_total))*100
percent_money = (num_money/float(num_total))*100
percent_cs = (num_cs/float(num_total))*100
percent_bag = (num_bag/float(num_total))*100
ntdata = [['Flight Cancellations and Delays', percent_flight], ['Price and Reimbursements', percent_money], ['Customer Service', percent_cs], ['Baggage and Handling', percent_bag]]
negative_tweets_df = pd.DataFrame(ntdata, columns = ['Issue', '% of Total Negative Tweets'])
negative_tweets_df.plot.bar(x='Issue', y='% of Total Negative Tweets', title='Negative Tweets by Issue Type')
plt.show()
```

```python
from collections import Counter

all_hashtags = []

def get_hashtags(tweet):
    """Return the hashtag"""
    return re.findall(r"#\w+", tweet)

for tweet in tweets:
    hashtags = get_hashtags(tweet)
    for hashtag in hashtags:
        all_hashtags.append(hashtag)
        
#Create a frequency counter for the hashtags. 
#This will allow you to count how many times each hashtag appears in the list.  
hashtag_counter = Counter(all_hashtags)

#Sort hashtags by frequency
sorted_hashtags = sorted(hashtag_counter.most_common(), key=lambda x: x[1], reverse=True)

#Print most common hashtags
for hashtag, count in sorted_hashtags[:15]:
    print(f"{hashtag}: {count}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create an instance of the Vader sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Replace 'airline_sentiment:confidence' with the actual column name
true_sentiment_column_name = 'airline_sentiment:confidence'

# Read the labeled dataset
labeled_data = pd.read_csv("D:/4th year/Grad Project/Datasets/Airline-Sentiment-2-w-AA.csv", encoding='ISO-8859-1')

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(labeled_data, test_size=0.2, random_state=42)

# Train the VADER model on the training set
train_sentiments = []
for tweet in train_data["text"]:
    train_sentiments.append(analyzer.polarity_scores(tweet).get("compound"))
train_data["predicted_sentiment"] = train_sentiments

# Use the trained model to predict sentiment on the testing set
test_sentiments = []
for tweet in test_data["text"]:
    test_sentiments.append(analyzer.polarity_scores(tweet).get("compound"))
test_data["predicted_sentiment"] = test_sentiments

# Convert compound scores to sentiment labels (positive, negative, neutral)
test_data["predicted_sentiment_label"] = test_data["predicted_sentiment"].apply(lambda x: "positive" if x > 0 else ("negative" if x < 0 else "neutral"))

# Convert true sentiment column to categorical labels
# For example, you can use a threshold to determine positive, negative, or neutral
threshold = 0.2  # Adjust the threshold as needed
test_data[true_sentiment_column_name + '_label'] = pd.cut(test_data[true_sentiment_column_name], bins=[-1, -threshold, threshold, 1], labels=['negative', 'neutral', 'positive'])

# Calculate accuracy
accuracy = accuracy_score(test_data[true_sentiment_column_name + '_label'], test_data["predicted_sentiment_label"])

print(f"Accuracy: {accuracy * 100:.2f}%")

# Display results in bar form
tweets = ['positive', 'negative', 'neutral']
result = [sum(test_data["predicted_sentiment_label"] == label) for label in tweets]
plt.bar(tweets, result)
plt.title('Sentiment Result')
plt.xlabel('Polarity')
plt.ylabel('Tweets')
plt.show()
```
